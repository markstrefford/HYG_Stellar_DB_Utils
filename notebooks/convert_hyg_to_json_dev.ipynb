# Development of convert_hyg_to_json.py

Notebook for developing conversion of HYG_v3 csv database to json file

```python
import os
import pandas as pd
import json
```

```python
HYG_DB_PATH = '../../HYG-Database'
HYG_DB_FILENAME = 'hygdata_v3.csv'

HYG_JSON_PATH = '../'
HYG_JSON_FILENAME = 'stars.json'
```

* Columns to convert to JSON

Column details are available at https://github.com/astronexus/HYG-Database

```python
JSON_COLUMNS = ['id', 'proper', 'ra', 'dec', 'dist', 
                'mag', 'absmag', 'spect', 'ci',
                'x', 'y', 'z', 'vx', 'vy', 'vz', 
                'lum', 'var', 'var_min', 'var_max']
```

* Load CSV and visualise

```python
hyg_db = pd.read_csv(os.path.join(HYG_DB_PATH, HYG_DB_FILENAME))[JSON_COLUMNS]
```

```python
hyg_db
```

* Convert to JSON

```python
result = hyg_db.to_json(orient='values')
parsed = json.loads(result)
# json.dumps(parsed, indent=4).split('\n')
```

```python
with open(os.path.join(HYG_JSON_PATH, HYG_JSON_FILE), 'w') as outfile:
    json.dump(parsed, outfile)
```
